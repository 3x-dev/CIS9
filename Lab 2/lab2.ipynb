#### B. Enrollment per school in the last 10 years

8. Do the following steps to find the average number of students for each academic year at each school.

8a. An academic year has 4 terms: Fall, Winter, Spring, Summer.<br>
Start with the cleaned DataFrame in step 6d, __create a new DataFrame with the average student count for the first academic year__.<br>
This means find the average student count of the first 4 columns, for each college, and store the average count in the new DataFrame with the column label 2013.<br>

First few lines of DataFrame:<br>
` 	                    2013`<br>
`Unnamed: 0 	            `<br>
`Alameda 	     5231.333333`<br>
`Allan Hancock 	13366.666667`<br>

```python
year_labels = ['Fall ' + str(year) for year in range(2013, 2024)]
academic_years_df = pd.DataFrame(index=clean_df.index)
```

8b. Continue with the rest of the columns:
- __find the average of every 4 columns__ (for the 4 terms of the year), but don't forget to stop when reaching Fall 2023, since this current academic year (2023-2024) is not done yet.
- __add the average to the DataFrame of step 8a, with the columnn label which is the year of the Fall term__.<br>
_You should not have to type the column label string, use the year of the Fall term instead_.

Then __print the first rows of the DataFrame__.

```python
for i in range(0, len(clean_df.columns) - 4, 4):  # Stop before Fall 2023
    year = year_labels[int(i/4)]
    academic_years_df[year] = clean_df.iloc[:, i:i+4].mean(axis=1)

academic_years_df.head()
```

9. Do the following steps to fnd the distribution of students per school for the last 10 years.

9a. Now that you have the average count of students for each year and for each school,<br>
__find the average student count for the last 10 years__ (2013-2022) and store in a Series.<br>
Then __print the Series__.

```python
average_student_count = academic_years_df.iloc[:, :10].mean(axis=1)
print(average_student_count)
```

9b. From the output of step 9a, __use an appropriate plot to show the distribution of average student count per school__.<br>
The plot should have: a title, x-axis label and y-axis label that explains what the x and y data are.<br>
Use a grid to make the plot easier to read.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
average_student_count.hist(grid=True, bins=30, rwidth=0.9)
plt.title('Distribution of Average Student Count per School (2013-2022)')
plt.xlabel('Average Student Count')
plt.ylabel('Number of Schools')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

9c. Use a pandas function to __show the basic statistics for the average student count per school__.

```python
statistics = average_student_count.describe()
print(statistics)
```

#### C. Enrollment trend of the largest colleges

10. Do the following steps to find the enrollment trend of the 8 colleges with the largest enrollment.

10a. __Sort the average student count__ and __print the top 8 colleges and their student counts__.

```python
top_8_colleges = average_student_count.sort_values(ascending=False).head(8)
print(top_8_colleges)
```

10b.  __Use an appropriate plot to show the trend of student counts in the past 10 years__ of the top 8 colleges.<br>

- The plot should have: a title, x-axis label and y-axis label that explains what the x and y data are, and with each college name clearly noted.<br>
- Use a grid to make the plot easier to read.
- Adjust the size of the plot so there's no overlapping of any features in the plot.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7)) 
for college in top_8_colleges.index:
    plt.plot(academic_years_df.columns, academic_years_df.loc[college], marker='o', label=college)

plt.title('Enrollment Trend for the Top 8 Colleges (2013-2023)')
plt.xlabel('Academic Year')
plt.ylabel('Average Student Count')
plt.legend(title='College Names', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### D. Recent change in enrollment for local community colleges

11. Do the following steps to find the percentage change in enrollment between Fall 2021 and Fall 2023, the most current year, for <u>local</u> community colleges.

11a. Using the cleaned DataFrame from step 6d, __find the percentage change in student counts between Fall 2021 and Fall 2023__ and store in a Series.<br>
Then use the given `colleges` list below to __find the percentage change for the local colleges__.

The local colleges are:<br>
`colleges = ['Deanza','Evergreen Valley','Foothill','Mission','San Jose City','West Valley']`

_Percentage change = (new value - old value) / old value * 100_

```python
percentage_change = ((clean_df['Fall 2023'] - clean_df['Fall 2021']) / clean_df['Fall 2021']) * 100
local_colleges_change = percentage_change.loc[['Deanza', 'Evergreen Valley', 'Foothill', 'Mission', 'San Jose City', 'West Valley']]
print(local_colleges_change)
```

11b. __Use an appropriate plot to show the enrollment change for the local colleges__.<br>
The plot should have: a title, x-axis label and y-axis label that explain what the x and y data are, and with each college name clearly noted.<br>
Use a grid to make the plot easier to read.

```python
plt.figure(figsize=(10, 6))
local_colleges_change.plot(kind='bar', color='red')
plt.title('Percentage Change in Enrollment for Local Colleges (2021-2023)')
plt.xlabel('College')
plt.ylabel('Percentage Change')
plt.axhline(0, color='grey', linewidth=0.8)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
